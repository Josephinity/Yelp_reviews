```python
import json
data = []
with open('yelp_academic_dataset_business.json') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
import pandas as pd
#yelp_academic_dataset_user.json
COLUMNS = data[0].keys()
df=pd.DataFrame({key:[d[key] for d in data[0:100]] for key in COLUMNS})
```

```python
def getKeySet(col):
    s=[]
    for i in range(0,len(col)):
        if(col[i]==None):
            a=[]
        else:
            a=col[i].keys()
        s = list(set(s)|set(a))
    return s
```

```python
def flattenDataFrame(df):
    for col in list(df.columns):
        if(type(df[col][0])==dict):
            sub_columns=getKeySet(df[col])
            for sub_col in sub_columns:
                df[col+'.'+sub_col]=[row[sub_col] if row is not None and sub_col in row.keys() else None for row in df[col]]
    return df
```

```python
flattenDataFrame(df)
df.to_csv('business.csv',encoding='utf-8')
```
